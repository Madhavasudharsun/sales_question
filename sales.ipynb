
** Identify average customer visit in the type B store in April Months***

```python
df = pd.merge(sales, stores, on='Store')

# Filter for Type B stores
type_b_stores = df[df['Type'] == 'B']

# Filter for sales data in April
april_data = type_b_stores[type_b_stores['Date'].dt.month == 4]

# Calculate average weekly sales (assuming weekly sales represent customer visits)
average_customer_visit = april_data['Weekly_Sales'].mean()

print(f'Average customer visit in Type B stores for April: {average_customer_visit}')
```
